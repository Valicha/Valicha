```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('listings.csv')
df.head()
```

```python
#Comprobamos que no existan valores nulos o valores en blanco en la tabla
df.isnull().sum()
```

```python

#Eliminamos las columnas que no vamos a necesitar
df.drop(['name','last_review','reviews_per_month','license','host_name'], axis=1)
```

```python

#Agruparemos por distrito y precio medio en forma descendente
df.groupby(["neighbourhood_group"])["price"].mean().sort_values(ascending=False)
```

```python

#Agruparemos por precio medio en cada barrio
df.groupby(["neighbourhood"])["price"].mean().sort_values(ascending=False)
```

```python

#Vamos a mostrar el top 10 de los más visitados según reviews
df.groupby(["neighbourhood"])["number_of_reviews"].sum().sort_values(ascending=False).head(10)
```

```python

#Mostraremos los barrios con mayor cantidad de Airbnb
# Agrupar por barrios, mostrar el distrito al que pertenecen, cuenta el número de id y ordenar por orden descendente
resultados = df.groupby(['neighbourhood', 'neighbourhood_group'])['id'].count().reset_index(name='Total').sort_values(['Total'], ascending=False).head(10)

# Imprime los resultados
print(resultados)
```

```python

# Mostraremos cantidad total de Airbnb por único propietario
#Agrupar por host_id y cuenta el número de veces que aparece en la columna
total_propietario = df.groupby(['host_id'])['host_id'].count().reset_index(name='Total').sort_values(['Total'], ascending=False)

# Imprime los resultados
print(total_propietario)
```

```python

#Mostraremos el total según tipos de habitaciones
# Agrupar por tipo de habitación y cuenta el número de veces que aparece en la columna
top_habitaciones = df.groupby(['room_type'])['room_type'].count().reset_index(name='Total').sort_values(['room_type'], ascending=False)

# Imprime los resultados
print(top_habitaciones)

```

```python

#Mostraremos el precio promedio según el tipo de habitación en cada barrio
# Agrupa por barrio y tipo de habitación, calcula el promedio del precio y ordena por orden descendente
resultados = df.groupby(['room_type','neighbourhood'])['price'].mean().reset_index(name='Precio_Medio').sort_values(['Precio_Medio'], ascending=False).head(10)

# Imprime los resultados
print(resultados)
```

```python

#Para finalizar, mostraremos la disponibilidad en promedio durante el año
# Agrupar por barrio y calcular el promedio de la disponibilidad, ordenar por orden descendente
resultados = df.groupby(['neighbourhood_group'])['availability_365'].mean().reset_index(name='Disponibilidad').sort_values(['Disponibilidad'], ascending=False)

# Imprime los resultados
print(resultados)
```
